```python

from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[4]").appName("test").getOrCreate()
```

```python
ls=[20,30,40,10,60]
rdd=spark.sparkContext.parallelize(ls)
```

```python
rdd.collect()
print("RDD count :",rdd.count())
print("RDD partition :",rdd.getNumPartitions())
print("RDD :", rdd.collect())
for i in rdd.collect():
    print(i)
```

```python
rddFile=spark.sparkContext.textFile("demo.txt")
rddFile.collect()
```

```python
rdd1=rddFile.flatMap(lambda x:x.split(" "))
#Flatmap returns the list of flattened elements
print("Flatmap output: ",rdd1.collect())
rddMap=rddFile.map(lambda x:x.split(" "))
#Map returns list of lists
print("Map output: ",rddMap.collect())
rdd2=rdd1.map(lambda x:(x,1))
rdd2.collect()
```

```python
rdd3=rdd2.reduceByKey(lambda x,y:x+y)
rdd3.collect()
'''reducebykey is better If you are grouping in order to perform an aggregation (such as a sum or average) over each key, 
using reduceByKey or aggregateByKey will provide much better performance.'''
rdd2.groupByKey().mapValues(len).collect()
```

```python
rdd4=rdd3.map(lambda x:(x[1],x[0])).sortByKey(False)
rdd4.collect()
```

```python
rdd3.sortBy(lambda x:x[1],False).collect()
```

```python
rdd3.takeOrdered(10,lambda x:x[1])
```

```python
rdd3.filter(lambda x:"a" in x[0][0]).collect()
```

```python
j=rdd4.join(rdd4)
j.collect()
```
